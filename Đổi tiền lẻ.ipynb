```python
count = 0
n = 0
for i in range(50+1):             #2k 50 tờ là đc 100k
    for j in range (33+1):        #3k 33 tờ là đc 100k
        for k in range(20+1):     #5k 20 tờ là đc 100k
            cash = i*2 + j*3 + k*5
            if cash == 100:
                print(f"Cách {n}: \t 2k {i} tờ \t 3k {j} tờ \t 5k {k} tờ")
                count += 1   
                n += 1
print (f"Có {count} cách quy đổi ")
```
