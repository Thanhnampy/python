```python
def vematcuoi():
    import turtle
    line = turtle.Turtle()
    line.pensize(5)
    line.speed(10)
    line.color("blue","bisque")

    #Vẽ khuôn mặt
    facesize = 100
    line.penup()
    line.goto(0, 100)
    line.pendown()
    line.begin_fill()
    line.circle(facesize)
    line.end_fill()
    
    #Vẽ thân
    bodysize = 200
    line.penup()
    line.goto(0, -300)
    line.pendown()
    line.begin_fill()
    line.circle(bodysize)
    line.end_fill()
    
    #vẽ mắt
    eye_sight = 15
    line.color("Black")
    
    line.begin_fill()
    line.penup()
    line.goto(-40,210)
    line.pendown()     
    line.circle(eye_sight)
    line.end_fill()
    
    line.begin_fill()
    line.penup()
    line.goto(40,210)
    line.pendown()   
    line.circle(eye_sight)
    line.end_fill()
    
    #Vẽ mũi 
    line.penup()
    line.goto(0,180)
    line.pendown()
    
    line.color("brown", "pink")
    line.begin_fill()
    line.circle(15, steps=3)
    line.end_fill()
    
    #Vẽ nụ cười
    line.penup()
    line.goto(-40,150)
    line.rt(90)
    line.color("black","yellow")
    line.begin_fill()
    
    line.pendown()
    line.circle(40,180)
    line.lt(90)
    line.fd(80)
    line.end_fill()
    
    

    turtle.done()
if __name__=="__main__":
    vematcuoi()
```
