```python
'''
Fizz Buzz là một bài toán lập trình cổ điển. Đây là phiên bản sửa đổi nhẹ của nó.
Viết chương trình lấy đầu vào là hai số nguyên: đầu và cuối khoảng (cả hai số đều thuộc khoảng).
Chương trình sẽ in ra các số trong khoảng này. Nhưng nếu số đó chia hết cho 3, 
bạn sẽ in ra “Fizz” thay vì nó, nếu số đó chia hết cho 5 in ra “Buzz” 
và nếu số đó chia hết cho cả 3 và 5, hãy in ra “FizzBuzz”.
'''
num1 = int(input("Start point =  "))
num2 = int(input("End point =  "))
i=num1
if num1>num2:
    print("Start point must lower than End point.")
else:
    for i in range (num1,num2 + 1):
        if i%3==0 and i%5==0: 
            print(f"{i}---fizzbuzz")
        elif i%5==0: 
            print(f"{i}---buzz")
        elif i%3==0: 
            print(f"{i}---fizz")
        else:
            print(f"{i}")
    i+=1
```
