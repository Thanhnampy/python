```python
#Bài 2: Nhập a, b, c là hệ số của một đa thức bậc 2. 
#    In ra phương trình bậc 2 dưới dạng chuẩn 
#    (dạng chuẩn có nghĩa là tự nhiên như khi viết tay đa thức ra giấy hoặc bảng). 
#    Các hệ số 1, -1 thì không in hệ số, 0 thì không in  bậc đó. 
#    Ví dụ: a = 1, b = -3, c= 0 thì in ra là x^2 – 3x.
```

```python
a = int(input("Nhập a = "))
b = int(input("Nhập b = "))
c = int(input("Nhập c = "))

if a == 0 and b == 0 :
    print("Invalid")
elif a == 1 and b == 1 and c > 0:
    print(f'Phương trình có dạng: x^2 + x + {c}')
elif a == 1 and b == 1 and c == 0:
    print(f'Phương trình có dạng: x^2 + x ')  
elif a == 1 and b == 1 and c < 0:
    print(f'Phương trình có dạng: x^2 + x {c}')
    
elif a == -1 and b == -1 and c > 0:
    print(f'Phương trình có dạng: -x^2 - x + {c}')
elif a == -1 and b == -1 and c == 0:
    print(f'Phương trình có dạng: -x^2 - x ')  
elif a == -1 and b == -1 and c < 0:
    print(f'Phương trình có dạng: -x^2 - x {c}')

elif a == 0 and b == 1 and c > 0:
    print(f'Phương trình có dạng:  x + {c}')
elif a == 0 and b == 1 and c == 0:
    print(f'Phương trình có dạng:  x ')  
elif a == 0 and b == 1 and c < 0:
    print(f'Phương trình có dạng:  x {c}')   
    
elif a == 1 and b == 0 and c > 0:
    print(f'Phương trình có dạng: x^2  + {c}')
elif a == 1 and b == 0 and c == 0:
    print(f'Phương trình có dạng: x^2 ')  
elif a == 1 and b == 0 and c < 0:
    print(f'Phương trình có dạng: x^2 {c}')


elif a == -1 and b == 1 and c > 0:
    print(f'Phương trình có dạng: -x^2 + x + {c}')
elif a == -1 and b == 1 and c == 0:
    print(f'Phương trình có dạng: -x^2 + x ')  
elif a == -1 and b == 1 and c < 0:
    print(f'Phương trình có dạng: -x^2 + x {c}')    

elif a == 1 and b == -1 and c > 0:
    print(f'Phương trình có dạng: x^2 - x + {c}')
elif a == 1 and b == -1 and c == 0:
    print(f'Phương trình có dạng: x^2 - x ')  
elif a == 1 and b == -1 and c < 0:
    print(f'Phương trình có dạng: x^2 - x {c}')
    
elif a == -1 and b == 0 and c > 0:
    print(f'Phương trình có dạng: -x^2  + {c}')
elif a == -1 and b == 0 and c == 0:
    print(f'Phương trình có dạng: -x^2  ')  
elif a == -1 and b == 0 and c < 0:
    print(f'Phương trình có dạng: -x^2 {c}')
    
elif (a != 1 or a != -1) and b == 0 and c > 0:
    print(f'Phương trình có dạng: {a}x^2  + {c}')
elif (a != 1 or a != -1) and b == 0 and c == 0:
    print(f'Phương trình có dạng: {a}x^2  ')  
elif (a != 1 or a != -1) and b == 0 and c < 0:
    print(f'Phương trình có dạng: {a}x^2 {c}')
    
    
elif a == 0 and b == -1 and c > 0:
    print(f'Phương trình có dạng: -x + {c}')
elif a == 0 and b == -1 and c == 0:
    print(f'Phương trình có dạng: -x ')  
elif a == 0 and b == -1 and c < 0:
    print(f'Phương trình có dạng: -x {c}')
    
elif a == 0 and (b != 1 or b != -1) and c > 0:
    print(f'Phương trình có dạng: {b}x + {c}')
elif a == 0 and (b != 1 or b != -1) and c == 0:
    print(f'Phương trình có dạng: {b}x ')  
elif a == 0 and (b != 1 or b != -1) and c < 0:
    print(f'Phương trình có dạng: {b}x {c}')

elif (a != 1 or a != -1) and (b != 1 or b != -1) and c > 0:
    print(f'Phương trình có dạng: {a}x^2 + {b}x + {c}')
elif (a != 1 or a != -1) and (b != 1 or b != -1) and c == 0:
    print(f'Phương trình có dạng: {a}x^2 + {b}x ')  
elif (a != 1 or a != -1) and (b != 1 or b != -1) and c < 0:
    print(f'Phương trình có dạng: {a}x^2 + {b}x {c}')

    

```
